```python
#printing prime numbers
for num in range(1, 51):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num)
```

```python
#printing even number
for num in range(1, 11):
  if(num%2==0):
    print("even numbers",num)
  else:
    print("odd numbers",num)

```
