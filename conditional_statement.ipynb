```python
#if-else statement
n=int(input("Enter number"))
if(n%2==0):
  print("Number is even")
else:
  print("Number is odd")
```

```python
#elif statement
n1=int(input("Enter first subject number"))
n2=int(input("Enter second subject number"))
n3=int(input("Enter third subject number"))
sum=n1+n2+n3
avg=sum/3
if(avg>=90):
  print("Grade A")
elif(avg>=80):
  print("Grade B")
elif(avg>=70):
  print("Grade C")
```

```python
#if statement
age=int(input("Enter your age"))
if(age>=18):
  print("You are eligible to vote")
else:
  print("You are not eligible to vote")
```
