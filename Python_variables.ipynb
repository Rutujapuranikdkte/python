```python
x=" awesome" #if declared global then it is accessible outside and inside the class
def myfunc():
  global x
  x="good"
myfunc()
print("python is:"+x)
```

```python
x = 'awesome'#not global then outside it prints awesome
def myfunc():
  x = 'fantastic'
  print('Python is ' + x)
myfunc()
print('Python is ' + x)
```
